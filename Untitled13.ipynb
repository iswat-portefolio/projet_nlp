```python
import os

# --- 1. Définition des chemins ---
files = {
    "business": "business.txt",
    "entertainment": "entertainment.txt",
    "tech": "tech.txt"
}

documents = []
labels = []

# --- 2. Charger chaque fichier ---
for label, filepath in files.items():
    if os.path.exists(filepath):
        with open(filepath, "r", encoding="utf-8") as f:
            text = f.read()
            documents.append(text)
            labels.append(label)
    else:
        print(f"❌ Fichier introuvable : {filepath}")

print("Nombre de documents chargés :", len(documents))
print("Labels :", labels)
```

```python
import re
import nltk
import spacy
from nltk.corpus import stopwords

nltk.download("stopwords")
stopwords_fr = set(stopwords.words("french"))

nlp = spacy.load("fr_core_news_md")

def clean_text(text):
    # Supprimer HTML
    text = re.sub(r"<.*?>", " ", text)
    # Supprimer URLs
    text = re.sub(r"http\S+|www\S+", " ", text)
    # Supprimer caractères spéciaux
    text = re.sub(r"[^a-zA-ZÀ-ÿ\s]", " ", text)
    # Minuscule
    text = text.lower()
    return text

def preprocess(text):
    text = clean_text(text)
    doc = nlp(text)

    tokens = []
    pos_tags = []
    dependencies = []

    for token in doc:
        if token.text not in stopwords_fr and len(token.text) > 2:
            tokens.append(token.lemma_)         # lemmatisation
            pos_tags.append((token.text, token.pos_))  # POS-tagging
            dependencies.append((token.text, token.dep_, token.head.text))  # Dépendances

    return tokens, pos_tags, dependencies
```

```python
processed_docs = []
pos_docs = []
dep_docs = []

for doc in documents:
    tokens, pos, deps = preprocess(doc)
    processed_docs.append(tokens)
    pos_docs.append(pos)
    dep_docs.append(deps)

print("Exemple tokens :", processed_docs[0][:20])
print("\nExemple POS :", pos_docs[0][:10])
print("\nExemple Dépendances :", dep_docs[0][:10])
```

```python
import spacy
nlp = spacy.load("fr_core_news_md")   # ou sm si md pas dispo

all_tokens = []
all_lemmas = []
all_pos = []
all_deps = []

for text in documents:      # <-- documents = liste de strings
    doc = nlp(text)

    tokens = [token.text for token in doc if not token.is_space]
    lemmas = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    pos_tags = [(token.text, token.pos_) for token in doc]
    deps = [(token.text, token.dep_, token.head.text) for token in doc]

    all_tokens.append(tokens)
    all_lemmas.append(lemmas)
    all_pos.append(pos_tags)
    all_deps.append(deps)
```

```python
print("TOKENS :", all_tokens[0][:30])
print("LEMMAS :", all_lemmas[0][:30])
print("POS :", all_pos[0][:10])
print("DEPS :", all_deps[0][:10])
```

```python
import pandas as pd

df = pd.DataFrame({
    "document": documents,
    "tokens": all_tokens,
    "lemmas": all_lemmas,
    "POS": all_pos,
    "dependencies": all_deps
})

df.head()
```
