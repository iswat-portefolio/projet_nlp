```python
#√âtape 2 : Cr√©er ton application Streamlit
%%writefile app.py
import streamlit as st
import pandas as pd

st.set_page_config(page_title="Assistant NLP BBC", layout="wide")

st.title("ü§ñ Assistant Intelligent de R√©sum√© et d'Extraction")
st.markdown("---")

# Zone de t√©l√©chargement
uploaded_file = st.file_uploader("Charger un document (TXT)", type="txt")

if uploaded_file:
    text = uploaded_file.read().decode("utf-8")

    col1, col2 = st.columns(2)

    with col1:
        st.subheader("Options d'analyse")
        model_type = st.selectbox("Mod√®le de classification", ["LSTM", "BERT", "DistilBERT"])
        if st.button("Lancer la classification"):
            st.info(f"R√©sultat : Th√©matique pr√©dite via {model_type}")

    with col2:
        st.subheader("Visualisation")
        st.text_area("Aper√ßu du texte", text[:1000], height=200)
```
